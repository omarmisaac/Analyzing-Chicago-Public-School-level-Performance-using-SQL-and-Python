
##### Number of Elementary Schools 

```python
%sql select count(*) from SCHOOLS where "Elementary, Middle, or High School" = 'ES'
```

#### The Highest Safety Score


```python
%sql select max(SAFETY_SCORE) AS MAX_SAFETY_SCORE from SCHOOLS;
```

##### The schools which have highest Safety Score


```python
%sql select NAME_OF_SCHOOL from SCHOOLS where Max(SAFETY_SCORE) from SCHOOLS);
```

##### The top 10 schools with the highest "Average Student Attendance"


```python
%sql select NAME_OF_SCHOOL from SCHOOLS order by AVERAGE_STUDENT_ATTENDANCE DESC NULLS LAST limit 10
```


##### The list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance


```python
%sql select REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%', '') from SCHOOLS order by AVERAGE_STUDENT_ATTENDANCE fetch first 5 rows only
```

##### Removing the '%' sign from the above result set for Average Student Attendance column


```python
%sql select REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%', '') from SCHOOLS order by AVERAGE_STUDENT_ATTENDANCE fetch first 5 rows only
```

##### The Schools who have Average Student Attendance lower than 70%?


```python
%sql select NAME_OF_SCHOOL from SCHOOLS WHERE AVERAGE_STUDENT_ATTENDANCE = SELECT (DECIMAL(REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%', ''))< 70)

#%sql SELECT Name_of_School, Average_Student_Attendance  \
     #from SCHOOLS \
    # where DECIMAL ( REPLACE(Average_Student_Attendance, '%', '') ) < 70 \
     #order by Average_Student_Attendance
```



#####  The total College Enrollment for each Community Area


```python
%sql select COMMUNITY_AREA_NAME, sum(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT from SCHOOLS group by COMMUNITY_AREA_NAME;
```



##### The 5 Community Areas with the least total College Enrollment  sorted in ascending order


```python
#%sql select COMMUNITY_AREA_NAME, sum(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT from SCHOOLS order by COLLEGE_ENROLLMENT ASC NULLS LAST LIMIT 5;

%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
   from SCHOOLS \
   group by Community_Area_Name \
   order by TOTAL_ENROLLMENT asc \
   fetch first 5 rows only
```


##### List of 5 schools with lowest safety score:


```python
%sql select NAME_OF_SCHOOL FROM SCHOOLS ORDER BY SAFETY_SCORE ASC LIMIT 5
```


#####  The hardship index for the community area which has College Enrollment of 4368


```python
%%sql 
select CD.HARDSHIP_INDEX 
   from chicago_socioeconomic_data CD, schools CPS 
   where CD.ca = CPS.community_area_number 
      and college_enrollment = 4368
```


##### The hardship index for the community area which has the school with the  highest enrollment.


```python
%sql select ca, community_area_name, hardship_index from chicago_socioeconomic_data \
   where ca = \
   ( select community_area_number from schools where college_enrollment = ( select max(college_enrollment) from SCHOOLS))
```
